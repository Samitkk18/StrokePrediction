```python
#Import the necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Load the dataset
dataset = pd.read_csv('healthcare-dataset-stroke-data.csv')
dataset.head()
```

```python
print(dataset['gender'].unique())
print(dataset['work_type'].unique())
print(dataset['Residence_type'].unique())
print(dataset['smoking_status'].unique())
```

```python
dataset.fillna(dataset["bmi"].median(), inplace=True)
```

```python
dataset.columns
```

```python
final_dataset = dataset[['id', 'gender', 'age', 'hypertension', 'heart_disease', 'ever_married',
       'work_type', 'Residence_type', 'avg_glucose_level', 'bmi',
       'smoking_status', 'stroke']]
```

```python
final_dataset=pd.get_dummies(final_dataset, drop_first=True)
```

```python
final_dataset.corr()
```

```python
corrmap = final_dataset.corr()
top_corr_features = corrmap.index
plt.figure(figsize=(20,20))
g=sns.heatmap(final_dataset[top_corr_features].corr(), annot=True)
```

```python
final_dataset.columns
```

```python
features = ['age', 'hypertension', 'heart_disease', 'avg_glucose_level',
       'bmi', 'gender_Male', 'gender_Other', 'ever_married_Yes',
       'work_type_Never_worked', 'work_type_Private',
       'work_type_Self_employed', 'work_type_children', 'Residence_type_Urban',
       'smoking_status_formerly_smoked', 'smoking_status_never_smoked',
       'smoking_status_smokes']
X = final_dataset[features]
y = final_dataset['stroke']
```

```python
X
```

```python
from imblearn.over_sampling import ADASYN
from collections import Counter
counter = Counter(y)
print('before :',counter)
ADA = ADASYN(random_state=130,sampling_strategy='minority')
X,y = ADA.fit_resample(X,y)
counter = Counter(y)
print("after :",counter)
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0,stratify=y)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
print(X_train,X_test)
```

```python
y_train
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,mean_absolute_error,confusion_matrix
nodes = [10,15,20,25,50,100,200]
accuracy =[]
accuracy1 =[]
for node in nodes:    
    criterion = ['entropy','gini']
    for criteria in criterion:
        random_model = RandomForestClassifier(criterion = criteria, random_state=0, n_estimators=node)
        random_model.fit(X_train,y_train)
        preds = random_model.predict(X_test)
        accs = accuracy_score(y_test,preds)
        if criteria == 'entropy':
            accuracy.append(accs)
        else:
            accuracy1.append(accs)
        print(confusion_matrix(y_test,preds),'No of Estimators: ', node,accs)

sns.lineplot(x=nodes,y=accuracy)
sns.lineplot(x=nodes,y=accuracy1)
```
